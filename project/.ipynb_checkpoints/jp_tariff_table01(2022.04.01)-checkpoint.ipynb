```python
#제1부(제1류) 관세율표 가져오기 - 2022년 4월 1일판
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from time import sleep

options= Options()
options.headless = True
driver = webdriver.Chrome(options = options)
url= "https://www.customs.go.jp/tariff/2022_04_01/data/print_j_01.htm"

driver.get(url)
sleep(1)

soup = BeautifulSoup(driver.page_source, 'lxml')
table = soup.find("table", attrs={"class":"t"}).find_all("td")
```

```python
# 33 x 81
print(len(table))

tables=[]
for i in range(len(table)):
    tables.append(table[i].get_text())
print(len(tables))
```

```python
#column 가져오기
columns=['H.S. code', '', 'Description', 'General', 'Temporary','WTO', 'GSP','LDC',
         'Singapore','Mexico','Malaysia','Chile','Thailand','Indonesia','Brunei',
         'ASEAN','Philippines','Switzerland','Viet Nam','India','Peru','Australia',
         'Mongolia','TPP11 (CPTPP)','EU','UK','RCEP(アセアン/豪州/ニュージーランド)',
         'RCEP(中国)','RCEP(韓国)','日米貿易協定','I','II','Law'
        ]
len(columns)
```

```python
import pandas as pd
import numpy as np
tables = np.reshape(tables,(81,33))
df = pd.DataFrame(tables, columns=columns)
df
```

```python
df.to_csv('japan_tariff_table01(2022.04.01).csv', mode='w', encoding='utf-8-sig')
```
