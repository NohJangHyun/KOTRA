```python
#과거 관세율표 전체 링크 수집
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from time import sleep

options= Options()
options.headless = True
driver = webdriver.Chrome(options = options)
url= "https://www.customs.go.jp/tariff/old_index.htm"

driver.get(url)
sleep(1)

soup = BeautifulSoup(driver.page_source, 'lxml')
temp= soup.find("div", attrs={"id":"contents"}).find_all('li')
```

```python
topics = []
links = []
i = 0
for li in temp:  
    a = li.find('a') 
    link = a['href']
    links.append("https://www.customs.go.jp/tariff/"+link)
    
    topic = a.get_text()
    topics.append(topic)
    
print(len(topics))
print(len(links))
```

```python
import pandas as pd

df = pd.DataFrame()
df['topic'] = topics
df['links'] = links
df
```

```python
df.to_csv('japan_past_tariff_link(2003~2021).csv', mode='w', encoding='utf-8-sig')
```
