```python
#METI(Ministry of Economoy, Trade and Industy) - 일본 경제산업성에서 올라오는 대외 경제 카테고리 뉴스

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from time import sleep

options= Options()
options.headless = True
driver = webdriver.Chrome(options = options)
url= "https://www.meti.go.jp/press/category_02.html"

driver.get(url)
sleep(1)

soup = BeautifulSoup(driver.page_source, 'lxml')
dl = soup.find("dl", attrs={"class":"date_sp b-solid"})
```

```python
#date
dates = []

dt = dl.find_all("dt")
for d in range(len(dt)):
    date = dt[d].get_text().replace("年", "/").replace("月","/").replace("日", "/")
    dates.append(date[:-1])
print(len(dates))
```

```python
#title & link
titles = []
links = []

A = dl.find_all("a")
for a in A:
    title = a.get_text()
    link = "https://www.meti.go.jp/"+a['href']
    titles.append(title)
    links.append(link)
print(len(titles))
print(len(links))
```

```python
#csv file
import pandas as pd

df = pd.DataFrame()
df['date'] = dates
df['title'] = titles
df['link'] = links
df
```

```python
df.to_csv('meti_released_news.csv', mode='w', encoding='utf-8-sig')
```

```python
#contents
contents = []
for link in links:
    print("url number",len(contents)+1, "is working")
    options= Options()
    options.headless = True
    driver = webdriver.Chrome(options = options)
    url= link
    
    driver.get(url)
    sleep(1)

    str_con = ""
    cont = []
    soup = BeautifulSoup(driver.page_source, 'lxml')
    content = soup.find("div", attrs={"class":"main w1000"})
    for con in content:
        c = con.get_text()
        if len(c) > 1:
            if c != "担当": 
                str_con = str_con + " " + c
            else:
                break
    cont.append(str_con.replace("\n", ""))
    contents.append(cont)
```

```python
#csv file
df['content'] = contents
df
```

```python
df.to_csv('meti_released_news_contents.csv', mode='w', encoding='utf-8-sig')
```
