```python
#전체 관세율표 합치기 - 2022년 4월 1일판
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd

options= Options()
options.headless = True
driver = webdriver.Chrome(options = options)
# df_all = pd.read_csv("japan_head.csv")
```

```python
for i in range(1,98):
    print(i)
    if i == 77:
        continue
    if i < 10:
        url = "https://www.customs.go.jp/tariff/2022_04_01/data/print_j_0"+str(i)+".htm"
    else:
        url = "https://www.customs.go.jp/tariff/2022_04_01/data/print_j_"+str(i)+".htm"
    
    driver.get(url)
    sleep(1)

    soup = BeautifulSoup(driver.page_source, 'lxml')
    table = soup.find_all("table")
    if i == 1:
        df_all = pd.read_html(driver.page_source)[2]
    else:
        df = pd.read_html(driver.page_source)[2]
        df_all = pd.concat([df_all, df])
```

```python
df_all.to_csv('japan_tariff_table_all(2022.04.01).csv', mode='w', encoding='utf-8-sig')
```
